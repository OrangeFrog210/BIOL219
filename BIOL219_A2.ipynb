```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
x = np.arange(0.01, 1000, 0.01)

p0 = [1/(1+math.exp(5+np.log(i/760))+(math.exp(-(2*(-5-np.log(i/760))-2.5)))) for i in x]
p1 = [math.exp(5+np.log(i/760))/(1+math.exp(5+np.log(i/760))+(math.exp(-(2*(-5-np.log(i/760))-2.5)))) for i in x]
p2 =[(math.exp(-(2*(-5-np.log(i/760))-2.5)))/(1+math.exp(5+np.log(i/760))+(math.exp(-(2*(-5-np.log(i/760))-2.5)))) for i in x]


plt.plot(x,p0, color='red')
plt.plot(x,p1, color='blue')
plt.plot(x,p2, color='green')

plt.xlim(0.01, 100)
plt.ylim(0, 1)
plt.xscale('log')
plt.xlabel('oxygen partial pressure (mmHg)', fontsize=12)
plt.ylabel('probability', fontsize=12)
```
